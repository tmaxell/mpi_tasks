```python
import matplotlib.pyplot as plt

# Данные
vector_sizes = [1000, 10000, 100000]
time_4_processes = [1.32046, 4.52371, 57.6867]
time_8_processes = [3.02608, 16.0358, 97.1361]

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(vector_sizes, time_4_processes, label="4 процесса", marker="o", color="blue")
plt.plot(vector_sizes, time_8_processes, label="8 процессов", marker="o", color="orange")

# Настройка осей
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Размер вектора")
plt.ylabel("Время выполнения (мс)")
plt.title("Зависимость времени выполнения от размера вектора")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Данные для построения графика
processes = ['Процесс 0', 'Процесс 1', 'Процесс 2', 'Процесс 3']
init_times = [0.058, 0.069, 0.061, 0.068]
comp_times = [2.245, 0.011, 2.244, 0.007]
total_times = [2.33, 0.082, 2.306, 0.075]

# Построение графика
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.2
index = range(len(processes))

bar1 = ax.bar(index, init_times, bar_width, label='Инициализация', color='b')
bar2 = ax.bar([p + bar_width for p in index], comp_times, bar_width, label='Вычисления', color='g')
bar3 = ax.bar([p + bar_width*2 for p in index], total_times, bar_width, label='Общее время', color='r')

ax.set_xlabel('Процессы')
ax.set_ylabel('Время (мс)')
ax.set_title('Время работы для каждого процесса')
ax.set_xticks([p + bar_width for p in index])
ax.set_xticklabels(processes)
ax.legend()

plt.tight_layout()
plt.show()
```
