```python
import matplotlib.pyplot as plt

# 
vector_sizes = [1000, 10000, 100000]
time_4_processes = [1.32046, 4.52371, 57.6867]
time_8_processes = [3.02608, 16.0358, 97.1361]

#  
plt.figure(figsize=(10, 6))
plt.plot(vector_sizes, time_4_processes, label="4 ", marker="o", color="blue")
plt.plot(vector_sizes, time_8_processes, label="8 ", marker="o", color="orange")

#  
plt.xscale("log")
plt.yscale("log")
plt.xlabel(" ")
plt.ylabel("  ()")
plt.title("     ")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

#    
processes = [' 0', ' 1', ' 2', ' 3']
init_times = [0.058, 0.069, 0.061, 0.068]
comp_times = [2.245, 0.011, 2.244, 0.007]
total_times = [2.33, 0.082, 2.306, 0.075]

#  
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.2
index = range(len(processes))

bar1 = ax.bar(index, init_times, bar_width, label='', color='b')
bar2 = ax.bar([p + bar_width for p in index], comp_times, bar_width, label='', color='g')
bar3 = ax.bar([p + bar_width*2 for p in index], total_times, bar_width, label=' ', color='r')

ax.set_xlabel('')
ax.set_ylabel(' ()')
ax.set_title('    ')
ax.set_xticks([p + bar_width for p in index])
ax.set_xticklabels(processes)
ax.legend()

plt.tight_layout()
plt.show()
```
